# Importing libraries.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
import seaborn as sns
```

# Preparing data.

## Importing data.

```python
#Training data csv import.
train_csv = pd.read_csv("./train.csv")

#Training data dataframe.
train_df = pd.DataFrame(train_csv)
```

## Label encoding.

```python
#Encoding textual categories into integral categories.

le = LabelEncoder()

train_df_encoded = train_df

train_df_encoded['job_description'] = le.fit_transform(train_df['job_description'])
print(le.classes_)

train_df_encoded['marital_status'] = le.fit_transform(train_df['marital_status'])
print(le.classes_)

train_df_encoded['education_details'] = le.fit_transform(train_df['education_details'])
print(le.classes_)

train_df_encoded['has_default'] = le.fit_transform(train_df['has_default'])
print(le.classes_)

train_df_encoded['housing_status'] = le.fit_transform(train_df['housing_status'])
print(le.classes_)

train_df_encoded['previous_loan'] = le.fit_transform(train_df['previous_loan'])
print(le.classes_)

train_df_encoded['phone_type'] = le.fit_transform(train_df['phone_type'])
print(le.classes_)

train_df_encoded['month_of_year'] = le.fit_transform(train_df['month_of_year'])
print(le.classes_)

train_df_encoded['poutcome_of_campaign'] = le.fit_transform(train_df['poutcome_of_campaign'])
print(le.classes_)

train_df_encoded['outcome'] = le.fit_transform(train_df['outcome'])
print(le.classes_)
```

**train_df_encoded** is the dataframe with integer encoded categorical values.

## Splitting features and target.

```python
#Train target.
y_train = train_df_encoded['outcome']

#Train features.
x_train = train_df_encoded.drop(axis = 0, columns=['outcome', 'serial_number'])
```
